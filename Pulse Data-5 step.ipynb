```python
import os
import glob
import pandas as pd
os.chdir("/data/localhost/taufiq/pulse-data-5-step")
```

```python
extension = 'final_.csv'
all_filenames = [i for i in glob.glob('*{}'.format(extension))]
pulse_data = pd.concat([pd.read_csv(f) for f in all_filenames ])
```

```python
pulse_data['icustay_id'] = pulse_data['id'].astype(int)
pulse_data = pulse_data.reset_index()
pulse_data.head()
```

```python
column_mappings = {
    'Albumin - BloodConcentration (mg/L)': 'Albumin',
    'ArterialCarbonDioxidePressure (mmHg)': 'paCO2',
    'ArterialOxygenPressure (mmHg)': 'paO2',
    'Bicarbonate - BloodConcentration (mg/L)': 'HCO3',
    'BloodPH (None)': 'Arterial_pH',
    'BloodUreaNitrogenConcentration (mg/dL)': 'BUN',
    'Calcium - BloodConcentration (mg/L)': 'Calcium',
    'Chloride - BloodConcentration (mg/L)': 'Chloride',
    'Creatinine - BloodConcentration (mg/L)': 'Creatinine',
    'DiastolicArterialPressure (mmHg)': 'DiaBP',
    'Glucose - BloodConcentration (mg/L)': 'Glucose',
    'Lactate - BloodConcentration (mg/L)': 'Arterial_lactate',
    'MeanArterialPressure (mmHg)': 'MeanBP',
    'Potassium - BloodConcentration (mg/L)': 'Potassium',
    'RespirationRate (1/min)': 'RR',
    'SaturationAndFractionOfInspiredOxygenRatio (None)': 'PaO2_FiO2',
    'SkinTemperature (degC)': 'Temp_C',
    'Sodium - BloodConcentration (mg/L)': 'Sodium',
    'SystolicArterialPressure (mmHg)': 'SysBP',
    'WhiteBloodCellCount (ct/uL)': 'WBC_count',
    'HeartRate (1/min)': 'HR'
}

pulse_rename = {}

for k, v in column_mappings.items():
    pulse_rename.update({k: f"{v}-pulse"})

pulse_data = pulse_data.rename(columns=pulse_rename)
```

```python
pulse_data_t4 = pulse_data[pulse_data['index']==4].reset_index(drop=True)
pulse_data_t4.head()
```

```python
MIMICtable = pd.read_csv("/data/localhost/taufiq/export-dir/MIMIC-1hourly-length-5.csv")
MIMICtable['icustay_id'] = MIMICtable['icustay_id'].astype(int)
MIMICtable.head()
```

```python
MIMICtable_filtered = MIMICtable[MIMICtable['icustay_id'].isin(pulse_data['icustay_id'])]
```

```python
MIMICtable_filtered_t0 = MIMICtable_filtered[MIMICtable_filtered['bloc']==1].reset_index()
MIMICtable_filtered_t4 = MIMICtable_filtered[MIMICtable_filtered['bloc']==4][[
       'icustay_id', 'RR', 'HR', 'SysBP', 'MeanBP', 'DiaBP', 'RR',
       'SpO2', 'Temp_C', 'FiO2_1', 'Potassium', 'Sodium', 'Chloride',
       'Glucose', 'BUN', 'Creatinine', 'Magnesium', 'Calcium', 'Ionised_Ca',
       'CO2_mEqL', 'SGOT', 'SGPT', 'Total_bili', 'Albumin', 'Hb', 'WBC_count',
       'Platelets_count', 'PTT', 'PT', 'INR', 'Arterial_pH', 'paO2', 'paCO2',
       'Arterial_BE', 'HCO3', 'Arterial_lactate']].reset_index()
```

```python
col_dict = {}
for col in MIMICtable_filtered_t4.columns:
    if col != 'icustay_id':
        col_dict.update({col: f"{col}-t4"})

MIMICtable_filtered_t4 = MIMICtable_filtered_t4.rename(columns=col_dict)
```

```python
data_combined = MIMICtable_filtered_t0.merge(MIMICtable_filtered_t4, left_on='icustay_id', right_on='icustay_id')
data_combined = data_combined.merge(pulse_data_t4, left_on='icustay_id', right_on='icustay_id')
```

```python
probs_data = pd.read_csv("/data/localhost/taufiq/export-dir/probability-data-5step.csv")
probs_data['icustay_id'] = probs_data['icustay_id'].astype(int)
```

```python
data_combined = data_combined.merge(probs_data, left_on='icustay_id', right_on='icustay_id')
```

```python
data_combined.head()
```

```python
grouped_max = pd.read_csv("/data/localhost/taufiq/export-dir/grouped_max.csv")
grouped_min = pd.read_csv("/data/localhost/taufiq/export-dir/grouped_min.csv")
for col in grouped_max:
    if col not in ['group_id']:
        grouped_max[f"{col}-max"] = grouped_max[col]
        grouped_max.drop(columns=[col], inplace=True)

for col in grouped_min:
    if col not in ['group_id']:
        grouped_min[f"{col}-min"] = grouped_min[col]
        grouped_min.drop(columns=[col], inplace=True)
data_combined = data_combined.merge(grouped_min, left_on='group_id', right_on='group_id')
data_combined = data_combined.merge(grouped_max, left_on='group_id', right_on='group_id')
data_combined.head()
```

```python
inbounds = 1
# for col in ['HR', 'SysBP', 'DiaBP', 'MeanBP', 'Sodium', 'Chloride', 'Glucose', 'Creatinine', 'Calcium']:
for col in ['HR', 'SysBP', 'DiaBP', 'MeanBP']:
# for col in ['Sodium', 'Chloride', 'Glucose', 'Creatinine']:    
    data_combined[f"{col}_ub"] = data_combined[col]*data_combined['prob'] + (1-data_combined['prob'])*(data_combined[f"{col}-max"])
    data_combined[f"{col}_lb"] = data_combined[col]*data_combined['prob'] + (1-data_combined['prob'])*(data_combined[f"{col}-min"])
    inbounds *= data_combined[f'{col}-pulse'].between(data_combined[f"{col}_lb"], data_combined[f"{col}_ub"])
```
