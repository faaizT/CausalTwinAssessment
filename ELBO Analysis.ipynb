```python
import pandas as pd
import seaborn as sns
import numpy as np
rho= [[0.0, 0.25], [0.5, 0.75]]
fig, axs = plt.subplots(2, 2, figsize=(12,12))
for i in range(2):
    for j in range(2):
        test_losses = pd.read_csv(f"/data/localhost/taufiq/model-learn-policy-posterior-rho_{rho[i][j]}-wd_0.1-lr_0.0001/outputs.csv")
        test_losses = test_losses.pivot_table(index="y", columns="x", values="test loss", aggfunc=np.min)
        img = sns.heatmap(test_losses, vmax=test_losses.values.max(), vmin=test_losses.values.min(), annot=True, fmt='.1f', ax=axs[i][j], cbar=False)
        axs[i][j].set_title(f"$\\rho$={rho[i][j]}")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
rho= [[0.0, 0.25], [0.5, 0.75]]
fig, axs = plt.subplots(2, 2, figsize=(12,12))
for i in range(2):
    for j in range(2):
        acc = pd.read_csv(f"/data/localhost/taufiq/model-learn-policy-posterior-rho_{rho[i][j]}-wd_0.1-lr_0.0001/accuracy.csv")
        acc = acc.pivot_table(index="y", columns="x", values="policy accuracy", aggfunc=np.min)
        img = sns.heatmap(acc, vmax=acc.values.max(), vmin=acc.values.min(), annot=True, fmt='.1f', ax=axs[i][j], cbar=False,  cmap=sns.cm.rocket_r)
        axs[i][j].set_title(f"$\\rho$={rho[i][j]}")
plt.show()
```
