```python
import pandas as pd
import seaborn as sns
import numpy as np
test_losses = pd.read_csv("/data/localhost/taufiq/model-wd-2/outputs.csv")
test_losses = test_losses.pivot_table(index="y", columns="x", values="test loss", aggfunc=np.min)
sns.heatmap(test_losses, vmax=test_losses.values.max(), vmin=test_losses.values.min(), annot=True, fmt='.1f')
```

```python
import pandas as pd
import seaborn as sns
from scipy.ndimage.filtersters import gaussian_filter
test_losses = pd.read_csv("/data/localhost/taufiq/simple-model-2/outputs.csv")
test_losses = test_losses.pivot("y", "x", "test loss")
loss_smooth = gaussian_filter(test_losses, sigma=0.50)
loss_smooth = pd.DataFrame(data=loss_smooth, index = test_losses.index, columns = test_losses.columns)
sns.heatmap(loss_smooth, vmax=loss_smooth.values.max(), vmin=loss_smooth.values.min(), annot=True, fmt='.0f')
```

```python
policy_acc = pd.read_csv("/data/localhost/taufiq/model-wd-2/accuracy.csv")
policy_acc = policy_acc.pivot_table(index="y", columns="x", values="policy accuracy", aggfunc=np.max)
sns.heatmap(policy_acc, vmax=policy_acc.values.max(), vmin=policy_acc.values.min(), annot=True, fmt='.0f', cmap=sns.cm.rocket_r)
```
