```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.cm as cm
def plot_loss_and_accuracy(model_name):
    fig, axs = plt.subplots(1, 2, figsize=(12,4))
    test_losses = pd.read_csv(f"/data/localhost/taufiq/{model_name}/outputs.csv")
    test_losses = test_losses.pivot_table(index="y", columns="x", values="test loss", aggfunc=np.min)
    sns.heatmap(test_losses, vmax=test_losses.values.max(), vmin=test_losses.values.min(), annot=True, fmt='.1f', ax=axs[0])
    policy_acc = pd.read_csv(f"/data/localhost/taufiq/{model_name}/accuracy.csv")
    policy_acc = policy_acc.pivot_table(index="y", columns="x", values="policy accuracy", aggfunc=np.max)
    sns.heatmap(policy_acc, vmax=policy_acc.values.max(), vmin=policy_acc.values.min(), annot=True, fmt='.0f', cmap=sns.cm.rocket_r, ax=axs[1])
    plt.show()
```

```python
import re
def plot_loss_and_accuracy_tensor(model_name):
    fig, axs = plt.subplots(1, 2, figsize=(12,4))
    test_losses = pd.read_csv(f"/data/localhost/taufiq/{model_name}/outputs.csv")
    loss = []
    for i in range(len(test_losses)):    
        m = re.search('tensor\((.+)\)', test_losses['test loss'][i])
        found = m.group(1)
        loss.append(float(found))
    test_losses.loc[:, 'test loss'] = loss
    test_losses = test_losses.pivot_table(index="y", columns="x", values="test loss", aggfunc=np.min)
    sns.heatmap(test_losses, vmax=test_losses.values.max(), vmin=test_losses.values.min(), annot=True, fmt='.2f', ax=axs[0])
    policy_acc = pd.read_csv(f"/data/localhost/taufiq/{model_name}/accuracy.csv")
    policy_acc = policy_acc.pivot_table(index="y", columns="x", values="policy accuracy", aggfunc=np.max)
    sns.heatmap(policy_acc, vmax=policy_acc.values.max(), vmin=policy_acc.values.min(), annot=True, fmt='.0f', cmap=sns.cm.rocket_r, ax=axs[1])
    plt.show()
```

```python
model_name = "model-learn-policy-posterior-wd_2.0-lr_0.0001"
plot_loss_and_accuracy(model_name)
```
