```python
import pandas as pd
df = pd.DataFrame()
for i in [5, 10, 15, 20, 25, 30, 35]:
    losses = pd.read_csv(f"/data/localhost/taufiq/simple-interventional-distribution/losses-{i}-20.csv")
    losses['Simulator'] = f'$g_{{{i},20}}$'
    df = df.append([df, losses], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(1, 1, figsize=(12,8))
sns.boxplot(x="action", y="loss", hue="Simulator", data=df[df['action']==1], palette="Set3", ax=ax, width=0.6)
ax.set_xlabel("Action",fontsize=20)
ax.set_ylabel("$-\mathcal{L}_{1}(X_{0:1})$",fontsize=20)
ax.tick_params(labelsize=20)
plt.legend(fontsize='x-large')
```

```python
df = pd.DataFrame()
for simulator in ['real', 'vaso2']:
    losses = pd.read_csv(f"/data/localhost/taufiq/gumbel-interventional-dist/losses-{simulator}.csv")
    if simulator == 'real':
        losses['Simulator'] = simulator
    else:
        losses['Simulator'] = '$vaso^2$'
    df = df.append([df, losses], ignore_index=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,8))
sns.boxplot(x="action", y="loss", hue="Simulator", data=df, palette="Set3", ax=ax, width=0.6)
ax.set_xlabel("Action",fontsize=20)
ax.set_ylabel("$-\mathcal{L}_{a}(X_{0:1})$",fontsize=20)
ax.tick_params(labelsize=20)
ax.set_ylim([0,9])
plt.legend(fontsize='x-large')
```
